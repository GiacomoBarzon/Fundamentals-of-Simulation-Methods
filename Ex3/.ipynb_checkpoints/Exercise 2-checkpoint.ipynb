### N-body problem, with N > 2 gravitating objects

### 1) Show that with an appropriate scaling $\vec x = \xi \tilde{\vec x}$, $\vec v = \phi \tilde{\vec v}$, $t = \tau \tilde t$ and  $m = \mu \tilde m$ (with $\xi$, $\phi$, $\tau$ and $\mu$ constants) the equations for N-body dynamics can be brought into dimensionless form

$\dfrac{d \vec{x_i}}{d t} = \xi \dfrac{d \tilde{\vec x_i}}{d t} = \vec v_i = \phi \tilde{\vec v_i}   $ 

### From now on we will do everything in dimensionless units

### 1) Write an N-body code for arbitrary N with the leapfrog integration and a constant time step. As a simple test problem solve a simple binary star problem. Plot the resulting trajectories of both stars in the (x, y)-plane (projection).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# initial conditions
x1 = np.array([-0.5,0.0,0.0])
v1 = np.array([0.0,-0.5,0.0])

x2 = np.array([0.5,0.0,0.0])
v2 = np.array([0.0,0.5,0.0])

m1 = 1.0
m2 = 1.0

x0 = np.array([x1,x2])
v0 = np.array([v1,v2])
m = np.array([m1,m2])

# integration parameters
T = 40.0 # total simulation time
dt = 0.001 # integration time-step
```

```python
# acceleration for a single object
def acci(x, m, i):
    xk = x - x[i,:]
    xk = np.delete(xk,i,0)
    xk = xk / (np.sum(xk**2,axis=1).reshape(x.shape[0]-1,1))**1.5
    xk = xk * np.delete(m, i).reshape(x.shape[0]-1,1)
    
    return np.sum(xk, axis=0)

# array of all accelerations vectors of the system
def acc(x, m):
    acc = np.zeros((x.shape[0],x.shape[1]))
    
    for i in range(x.shape[0]):
        acc[i] = acci(x, m, i)
    
    return acc

# LEAP-FROG ALGORITHM
def LeapFrog(x0, v0, dt):
    N = x0.shape[0] # number of objects
    steps = np.int(np.ceil(T/dt)) # total number of steps
    
    xt = np.zeros((x0.shape[0], x0.shape[1], steps+1))
    vt = np.zeros((v0.shape[0], v0.shape[1], steps+1))
    
    xt[:,:,0] = x0
    vt[:,:,0] = v0
    
    # loop over steps
    for i in range(1,steps+1):
        vhalf = vt[:,:,i-1] + acc(xt[:,:,i-1], m) * dt / 2.
        xt[:,:,i] = xt[:,:,i-1] + vhalf * dt
        vt[:,:,i] = vhalf + acc(xt[:,:,i], m) * dt / 2.
    
    return (xt, vt)
    
```

```python
# Running Leap-Frog algorithm
(Xt, Vt) = LeapFrog(x0, v0, dt)
```

```python
# Plot results
plt.figure(figsize=[12,8])
plt.plot(Xt[0,0,:], Xt[0,1,:], label=r'$x_{1}$')
plt.plot(Xt[1,0,:], Xt[1,1,:], label=r'$x_{2}$')
plt.plot(Xt[0,0,0], Xt[0,1,0],'o', label=r'initial $x_{1}$')
plt.plot(Xt[1,0,0], Xt[1,1,0],'o', label=r'initial $x_{2}$')
plt.xlabel('x', fontsize = 20)
plt.ylabel('y', fontsize = 20)
plt.legend()
```

### 2) Now add a third star. Show how this third star “falls into” the binary, interacts with it, and gets eventually ejected. Plot the trajectories of all three stars in the (x, y)-plane (projection)

```python
# new initial conditions
x3 = np.array([1.0,6.0,2.0])
v3 = np.array([0.0,0.0,0.0])
m3 = 0.1

x0 = np.append(x0, [x3], axis=0)
v0 = np.append(v0, [v3], axis=0)
m = np.append(m, m3)
```

```python
# Running Leap-Frog algorithm
(Xt, Vt) = LeapFrog(x0, v0, dt)
```

```python
# Plot results
for i in range(Xt.shape[0]):
    plt.plot(Xt[i,0,:], Xt[i,1,:])
```
