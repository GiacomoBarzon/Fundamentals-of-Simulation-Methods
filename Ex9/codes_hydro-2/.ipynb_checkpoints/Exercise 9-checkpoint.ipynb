# 1.1. Isothermal 1D hydrodynamics solver

### 1. Based on th egeneral-purpose advection routine of the previous exercise construct a simple isothermal hydrodynamics solver in 1D. Implement periodic boundary conditions using the ghost cell technique.

```python
import numpy as np
import matplotlib.pyplot as plt 
from hydro_iso_classic import *
from hydro_iso_riemann import *
from bc import *
```

```python
# define parameters
cs = 1.

x_min = -100.
x_max = 100.
N = 200 # number of cells
dx = (x_max-x_min)/(N-1)

T = 100
dt = 0.05
steps = int(T / dt) +1

# adding two ghost cells
x = np.arange(x_min - dx, x_max + dx, dx)

# define initial conditions
rho_0 = 1 + np.exp( -x**2 / 200.)
u_0 = np.zeros(x.shape[0])

q_0 = np.array(( [rho_0, rho_0*u_0] ))
```

```python
def CFL(dx, q):
    v_max = max(abs(q[1]/q[0]))
    return dx / v_max

def iso_integrator(q_0, cs, dx, dt, steps):
    q = np.zeros(( [steps+1, q_0.shape[0], q_0.shape[1]] ))
    
    # set initial conditions
    q[0] = q_0
    
    # loop over timesteps
    for i in range(steps):
        q[i+1] = hydro_iso_classic_one_timestep(q[i],cs,dx,dt)
        
        # implement periodic boundary conditions
        implement_boundcond(q[i+1],bctype='periodic')
        
        # check CFL condition
        if not ( dt < CFL(dx, q[i+1]) ):
            print('CFL timestep condition not satisfied at timestep %i' %(i+1) )
        
    return q
```

### 2. Use your code to solve the following 1D isothermal hydrodynamics problem: the x-grid goes from x=âˆ’100 to x=100, the boundary conditions are periodic, the isothermal sound speed is cs = 1. The initial conditions are:
$$ \rho(x,t=0) = 1 + \exp(-x^2/200)  $$
$$ u(x,t=0) = 0 $$

```python
q = iso_integrator(q_0, cs, dx, dt, steps)
```

### Plot the system at 10 different times: t = 10, t = 20, ..., t = 90, t = 100. Describe your results, and try to explain what you see.

```python
ts = np.arange(0,T+1,10) / dt
ts = ts.astype(int)

# plot solution
fig1 = plt.subplots(figsize=[15,12])

plt.subplot(2,1,1)
for t in ts:
    plt.plot(x[1:-1], q[t,0,1:-1], label='t=%.0f' %(t*dt) )
plt.xlabel('x', fontsize = 18)
plt.ylabel(r'$\rho(x,t)$', fontsize = 18)
plt.grid()
plt.legend()
plt.title('Density profile', fontsize=20)

plt.subplot(2,1,2)
for t in ts:
    plt.plot(x[1:-1], q[t,1,1:-1] / q[t,0,1:-1], label='t=%.0f' %(t*dt) )
plt.xlabel('x', fontsize = 18)
plt.ylabel(r'$v(x,t)$', fontsize = 18)
plt.grid()
plt.legend()
plt.title('Velocity profile', fontsize=20)

plt.show(fig1)
```

### 3. Now do the same, but with variable time step. Calculate the CFL condition $\Delta t_{CFL}$ at each time step and choose $\Delta t$ = 0.4 * $\Delta t_{CFL}$ for safety

```python
print_dt = 0.5

def iso_integrator_variable(q_0, cs, dx, T, inital_dt=0.1):
    # set initial conditions
    q = np.copy(q_0)
    q = q.reshape(1,q_0.shape[0],q_0.shape[1])
    ts = np.zeros(1)
    
    new_dt = inital_dt
    ts = np.append(ts, new_dt)
    i = 1
    
    # loop over timesteps
    while ts[-1] <= T:
        q = np.vstack(( q, hydro_iso_classic_one_timestep(q[-1],cs,dx,new_dt).reshape(1,q_0.shape[0],q_0.shape[1]) ))
        
        # implement periodic boundary conditions
        implement_boundcond(q[-1],bctype='periodic')
        
        # compute new timestep
        new_dt = 0.4 * CFL(dx, q[-1])
        
        # check that have solution almost every print_dt
        if (ts[-1] + new_dt < i*print_dt):
            ts = np.append( ts, ts[-1] + new_dt )
        else:
            ts = np.append(ts, i*print_dt)
            new_dt = ts[-1] - ts[-2]
            i += 1
        
    return q, np.array( ts[:-1] )
```

```python
q, times = iso_integrator_variable(q_0, cs, dx, T)
print('Total number of timesteps: ', times.shape[0])
```

```python
# plot solution
fig1 = plt.subplots(figsize=[15,12])

t_toprint = times[times%10 == 0]
q_toprint = q[times%10 == 0, :, :]

plt.subplot(2,1,1)
for i,t in enumerate(t_toprint):
    plt.plot(x[1:-1], q_toprint[i,0,1:-1], label='t=%.1f' %(t) )
plt.xlabel('x', fontsize = 18)
plt.ylabel(r'$\rho(x,t)$', fontsize = 18)
plt.grid()
plt.legend()
plt.title('Density profile', fontsize=20)

plt.subplot(2,1,2)
for i,t in enumerate(t_toprint):
    plt.plot(x[1:-1], q_toprint[i,1,1:-1]/q_toprint[i,0,1:-1], label='t=%.1f' %(t) )
plt.xlabel('x', fontsize = 18)
plt.ylabel(r'$v(x,t)$', fontsize = 18)
plt.grid()
plt.legend()
plt.title('Velocity profile', fontsize=20)

plt.show(fig1)
```

# 1.2. HLL Riemann solver

### 1. Apply the Riemann solver to the above wave problem, and show that the results are better, i. e. have less spurious oscillations

```python
# Define function to integrate over time
def riemann_integrator(q_0, cs, dx, dt, steps):    
    # create empty array
    q = np.zeros(( [steps+1,q_0.shape[0],q_0.shape[1]] ))
    
    # set initial conditions
    q[0] = q_0

    # loop over time steps
    for i in range(steps):
        q[i+1] = hydro_iso_riemann_one_timestep(q[i],cs,dx,dt)
        
        # check CFL condition
        if not ( dt < CFL(dx, q[i+1]) ):
            print('CFL timestep condition not satisfied at timestep %i' %(i+1) )
        
        # implement periodic boundary conditions
        implement_boundcond(q[i+1],bctype='periodic')
        
        # check CFL condition
        if not ( dt < CFL(dx, q[i+1]) ):
            print('CFL timestep condition not satisfied at timestep %i' %(i+1) )
        
    return q
```

```python
q = riemann_integrator(q_0, cs, dx, dt, steps)
```

```python
# plot solution
fig1 = plt.subplots(figsize=[15,12])

plt.subplot(2,1,1)
for t in ts:
    plt.plot(x[1:-1], q[t,0,1:-1], label='t=%.0f' %(t*dt) )
plt.xlabel('x', fontsize = 18)
plt.ylabel(r'$\rho(x,t)$', fontsize = 18)
plt.grid()
plt.legend()
plt.title('Density profile', fontsize=20)

plt.subplot(2,1,2)
for t in ts:
    plt.plot(x[1:-1], q[t,1,1:-1] / q[t,0,1:-1], label='t=%.0f' %(t*dt) )
plt.xlabel('x', fontsize = 18)
plt.ylabel(r'$v(x,t)$', fontsize = 18)
plt.grid()
plt.legend()
plt.title('Velocity profile', fontsize=20)

plt.show(fig1)
```
