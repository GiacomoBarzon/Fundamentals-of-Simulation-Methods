```python
import numpy as np
import math
from math import exp
import matplotlib.pyplot as plt
```

# Defining function and asking for value

```python
def f(x):
    return (x+np.exp(-x)-1.)/(x**2)

val = input("Value of x: ")
print("f(x =",float(val),")=",f(float(val)))
```

```python
epsilonx = 1E-6
epsilony = 0.0

lin = np.arange(0.0, 1E-4, epsilonx)
x_wrong = lin[1]
corr = np.zeros(len(lin))
for i in range(len(lin)-1,1,-1):
    if (f(lin[i])>(f(lin[i-1])+epsilony) or f(lin[i])>(0.5+epsilony)):
    #if  abs(f(lin[i])-0.5)>epsilony:
        x_wrong = lin[i]
        print("The evaluation goes wrong at x = ", lin[i])
        break
        
plt.figure(figsize=[15,10])
plt.plot(lin, f(lin), 'o')
plt.plot(lin, np.zeros(len(lin))+0.5)
plt.axvline(x=x_wrong, linestyle = '--')
plt.xlabel('x', fontsize = 20)
plt.ylabel('f(x)', fontsize = 20)
plt.title('Plotting of f(x)', fontsize = 20)
plt.show()
```

```python
def new_f(x):
    return 0.5 - x/6
```

```python
t1 = np.arange(0.0, x_wrong+epsilonx, epsilonx)
t2 = np.arange(x_wrong+epsilonx, 1E-4, epsilonx)

plt.figure(figsize=[15,10])
plt.plot(t1, new_f(t1), 'o')
plt.plot(t2, f(t2), 'o')
plt.plot(lin, np.zeros(len(lin))+0.5)
plt.axvline(x=x_wrong, linestyle = '--')
plt.ylim([0.4999, 0.5001])
plt.xlabel('x', fontsize = 20)
plt.ylabel('f(x)', fontsize = 20)
plt.title('Corrected behaviour of f(x)', fontsize = 20)
plt.show()
```

```python
def correct_f(x):
    if x > x_wrong:
        return (x+np.exp(-x)-1.)/(x**2)
    else:
        return 0.5 - x/6.
    
correct_f = np.vectorize(correct_f)
```

```python
t = np.arange(0.0, 1E-4, epsilonx)

plt.figure(figsize=[15,10])
plt.plot(t,f(t), '.')
plt.plot(t,correct_f(t),'.')
plt.plot(lin, np.zeros(len(lin))+0.5)
plt.axvline(x=x_wrong, linestyle = '--')
plt.xlabel('x', fontsize = 20)
plt.ylabel('y', fontsize = 20)
plt.title('Comparison of the two functions', fontsize = 20)
plt.legend(['f(x)','correct_f(x)','lim f(x) x->0','x_wrong'])
plt.show()
```
